```python
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv("IBM_HR_ATTRITION.csv")
dataset.tail(5)
```

```python
dataset["Attrition"].unique()
```

```python
dataset['Attrition'] = dataset['Attrition'].apply(lambda x: 1 if x.lower()=="yes" else 0)
```

```python
import matplotlib.pyplot as plt
sns.lmplot(x="DistanceFromHome", y="Attrition", data= dataset, logistic = True, scatter_kws={"alpha":.05})
plt.show()
```

#### Seaborn's lmplot function was used to create a logistic regression plot between the variables "DistanceFromHome" and "Attrition" in the dataset , let's analyse this:
#### The graph shows a linear relationship between distance from home and attrition, with attrition increasing as distance from home increases. This suggests that employees who live further away from their workplaces are more likely to leave their jobs.
#### It is important to note that the correlation between distance from home and attrition does not necessarily mean that distance from home causes attrition. There could be other factors that are correlated with both distance from home and attrition, such as age, experience, or salary...

```python
sns.lmplot(x="YearsAtCompany", y="Attrition", data= dataset, logistic = True)
plt.show()
```

#### The graph shows the attrition rate of employees at a company over time. The graph shows a clear trend of increasing attrition rate over time ,  this means that employees who have been with a company for longer are more likely to stay than employees who have been with the company for a shorter period of time. Maybe the ones who spent lots of time in the company are satistfied with the salary and their missions in the company unlike the ones who stayed for a shorter time.

```python
sns.lmplot(x="Age", y="Attrition", data= dataset, logistic = True)
plt.show()
```

#### 
the graph indicates that younger employees are more inclined to leave the company for various reasons: searching for other opportunities or seeking higher salaries quickly by changing companies. Additionally, as age decreases, the attrition rate also decreases.
