```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("IBM_HR_ATTRITION.csv")
dataset.head()
```

###  Constructing a graph that allows us to check if the distribution JobSatisfaction is different for people who left the company compared to people who didn't 

```python
sns.catplot(x="JobSatisfaction", y="Attrition", data = dataset, kind= "box")
plt.show()
```

#### The job satisfaction score goes from 1 to 4 . We notice that there are some employees who have a low job satisfaction ( from 1 to 2) but did not quit their job . We also notice that 50% of the employees who did not quit the company have a job satisfaction from 2 to 4.
#### Let's move to the ones who chose to quit IBM company, we notice that 50% of them have a score of job satisfaction from 1 to 3 , a group of them have a job satisfaction score from 3 to 4 but still make the choice to quit their jobs.

```python
sns.catplot(x="TotalWorkingYears", y="Attrition", data = dataset, kind= "box")
plt.show()
```

#### We notice the presence of outliers in both boxplots. 
#### Let's have a look on the boxplot of people who did not quit their jobs. We notice that 50% of them worked from 6 to 16 years , the range of data goes from 0 to 31 working years (without outliers) . The data is right skewed because the right tail in long .
#### Let's have a look on the boxplot of people who did quit their jobs. We notice that 50% of them worked from 4 to 10 years , the range of data goes from 0 to 20 working years (without outliers). The data is right skewed because the right tail in long .
