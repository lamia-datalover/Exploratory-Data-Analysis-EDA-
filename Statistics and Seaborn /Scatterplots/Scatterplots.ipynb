```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("house_pricing.csv")
dataset.head()
```

```python
dataset.info()
```

### Bhavior of LotArea column in relation to the SalePrices column

```python
sns.relplot(x="SalePrice", y="LotArea", data=dataset)
```

### Using this visual representation it is hard to hard to get valuable information let's refine it by keeping only houses that have a LotArea less than 20000 square feet and a price less than 500,000 dollar

```python
sns.relplot(x="LotArea", y="SalePrice", 
            data=dataset.loc[(dataset['LotArea'] < 20000) & (dataset['SalePrice'] < 500000), :])
plt.show()
```

#### The scatter plot  shows that there is a lot of variation in the data. Some houses with large lots have sold for relatively low prices, while other houses with small lots have sold for relatively high prices. This variation is likely due to a number of factors, such as the condition of the house, the location of the house, and the features of the house

###  The relationship between the LotFrontage and the LotArea

```python
sns.relplot(x="LotFrontage", y="LotArea", data=dataset)
plt.show()
```

### Refining the last visualization by keeping only those houses that have a LotFrontage of less than 200 square feet and a LotArea of less than 100,000 square feet.

```python
sns.relplot(x="LotFrontage", y="LotArea", data=dataset[(dataset.LotFrontage < 200) &
                                                       (dataset.LotArea < 100000)])
plt.show()
```

#### The scatter plot shows a positive correlation between lot size and lot frontage. This means that larger lots tend to have longer frontages. 
#### There are a few outliers in the data. These are lots that have much shorter or longer frontages than expected for their size.
#### The relationship between lot size and lot frontage is not perfectly linear. This means that there is some scatter in the data. 
#### ( Scatter in data refers to the random spread of data points around the trendline. In other words, it is the variability of the data points. The more scatter there is in the data, the less linear the relationship between the two variables, there are a few reasons why there might be scatter in data. One reason is that the two variables may not be perfectly correlated. This means that there may be other factors that affect the dependent variable besides the independent variable. Another reason for scatter is that there may be measurement error in the data. This means that the data points may not be accurate reflections of the true values of the variables)
