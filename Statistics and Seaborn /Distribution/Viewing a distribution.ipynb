```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("IBM_HR_ATTRITION.csv")
```

```python
sns.distplot(dataset['MonthlyIncome'])
```

### We can see that the very high salaries bias our distribution, let's try to see the distribution of salaries between 0 and 8000 dollar per month, the image shows a normalized distribution of the Monthly Income of employees in IBM company, with a slight skew towards the right.This means that the high salaries are in the right we should dig more !

```python
sns.distplot(dataset.loc[dataset['MonthlyIncome'] < 8000, 'MonthlyIncome'], bins = 40)
```

#### This image displays the salary distribution ranging from 0 to 8000 dollars. It's evident that the income of 2500 dollars has the highest density at 0.0028. Following this value, the curve sharply declines to 0.00016 before gradually increasing to 0.0018. Subsequently, it continues to decrease gradually. 
#### The shape of the curve in a this graph indicates the concentration of data points. Peaks in the curve suggest where values are more densely clustered, while lower areas indicate less frequent values. A wider or flatter curve signifies a more dispersed or spread-out distribution, while a narrower and taller curve suggests a more concentrated distribution.
