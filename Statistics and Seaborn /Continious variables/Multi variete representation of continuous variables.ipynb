```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("sales_predictions.csv")
dataset.head()
```

```python
#Working on a sample of 25 random entries from the dataset
dataset = dataset.sample(25)
dataset.head()
```

```python
sns.relplot(x="date", y="item_price", data = dataset, kind="line", height = 5, aspect = 5)
plt.show()
```

### The results are not very visible let's convert the date column to datetime 

```python
dataset['date'] = pd.to_datetime(dataset['date'])
```

```python
sns.relplot(x="date", y="item_price", data = dataset, kind="line", height = 5, aspect = 5)
plt.show()
```

#### We observed that the provided data showed no evident seasonality.
#### From 2013 to September 2014-08, maintaining a consistent item price with occasional fluctuations. Notably, there was a price surge in July 2013-07 and another increase in April 2014. However, in October 2014, there was a sudden sharp rise, reaching a price of 1800. Subsequently, it drastically dropped to 400, then unexpectedly spiked to 3100 in 2015-01. By May 2015, it settled at a constant value of 750.
